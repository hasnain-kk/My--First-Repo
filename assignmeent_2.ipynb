Q 1) **What is a dashboard in data visualization, and what are its key components?**

   A dashboard in data visualization provides a clear and interactive way to monitor important metrics, allowing users to track performance or analyze data patterns in real-time. It typically contains several visual components like charts, graphs, and tables, making data insights accessible.

   **Key Components:**
   - **Charts & Graphs:** To represent data trends and relationships.
   - **Filters:** Enable users to drill down into specific data points.
   - **KPIs:** Key performance indicators to measure performance.
   - **Real-Time Updates:** Often provides live or near-real-time data insights.

```python
# Dataset
data = [
    ['Product', 'Sales_Q1', 'Sales_Q2', 'Sales_Q3', 'Sales_Q4', 'Profit_Margin', 'Region'],
    ['A', 150, 200, 250, 300, 3, 'North'],
    ['B', 300, 350, 400, 450, 15, 'East'],
    ['C', 100, 120, 180, 210, 8, 'West'],
    ['D', 200, 380, 500, 520, 12, 'South'],
    ['E', 250, 270, 300, 330, 20, 'North']
]

import pandas as pd

# Create a DataFrame
df = pd.DataFrame(data[1:], columns=data[0])
df
```

Q 3:) **Plot a line chart to show the sales trend of all products across the four quarters. Each product should be represented by a different line. Include appropriate labels, title, and legend.**

**Hints:**
- Use `plt.plot()` for line plots.
- Use `plt.legend()` to display the product names.

```python
import matplotlib.pyplot as plt

# Plotting the sales trend
plt.figure(figsize=(10, 6))
for product in df['Product']:
    plt.plot(['Q1', 'Q2', 'Q3', 'Q4'], df[df['Product'] == product].iloc[0, 1:5], label=product)

plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Sales Trend Across Quarters')
plt.legend(title='Product')
plt.grid(False)
plt.show()
```

Q 4) **Create a bar plot to show the total sales (sum of all four quarters) of each product, grouped by region. Use color coding for regions and annotate the bars with total sales values.**

**Hints:**

Use sns.barplot() with hue for region.
Add annotations with plt.text().

```python
import seaborn as sns

# Calculate total sales for each product
df['Total_Sales'] = df[['Sales_Q1', 'Sales_Q2', 'Sales_Q3', 'Sales_Q4']].sum(axis=1)

# Plotting the bar plot
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='Product', y='Total_Sales', hue='Region', data=df, palette='viridis')

# Annotating the bars with total sales values
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.0f'), 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha = 'center', va = 'center', 
                      xytext = (0, 9), 
                      textcoords = 'offset points')

plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales of Each Product Grouped by Region')
plt.show()
```

Q 5) **enerate a scatter plot to show the relationship between total sales (sum of Q1 to Q4) and profit margin for each product. Differentiate the points by product and region using color and markers.**

**Hints:**

Use sns.scatterplot() with hue for product and style for region.

```python
# Plotting the scatter plot
plt.figure(figsize=(12, 6))
scatter_plot = sns.scatterplot(x='Total_Sales', y='Profit_Margin', hue='Product', style='Region', data=df, palette='tab10')

plt.xlabel('Total Sales')
plt.ylabel('Profit Margin')
plt.title('Total Sales vs. Profit Margin')
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
